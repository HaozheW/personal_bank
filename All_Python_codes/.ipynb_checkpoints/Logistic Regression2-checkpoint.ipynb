```python
import collections
import pandas as pd
import numpy as np
from datetime import datetime
import os
import statsmodels.api as sm
import pylab as pl
from sklearn.cluster import KMeans

pd.set_option('display.max_columns', 100)#this allowed 80 columns to be shown in ipython notebook
```

```python
# read data into pandas dataframe and convert column names to upper case
def read_data(pname,fname):
	if fname.endswith('.csv'):
		df = pd.DataFrame.from_csv(os.path.join(pname,fname),index_col=None)
		df = df.rename(columns=lambda x: x.upper(),inplace=False)
		return df
	elif fname.endswith('.xlsx'):
		df = pd.io.excel.read_excel(os.path.join(pname,fname),sheetname=0,\
			index_col=None)
		df = df.rename(columns=lambda x: x.upper(),inplace=False)
		return df




# read data into pandas dataframes
audit = read_data(pathname,"newsample_0722.csv")
print "total number of audits:", len(audit)
print "Data read into dataframes"
```

```python
audit.insert(0, 'CONVERT', audit['CLASS'])
del audit['CLASS']
audit.insert(1, 'SQFT_LN', np.log(audit['SQFT']))
audit.insert(2, 'ANNUALPRIMARYUSAGE_LN', np.log(audit['ANNUALPRIMARYUSAGE']))
audit.insert(3, 'MEDIANHOUSEVALUE_LN', np.log(audit['MEDIANHOUSEVALUE']))
audit = audit.dropna()

```

```python
count = audit.groupby(['AUDITCONTRACTOR']).count().sort('CONVERT')
count.insert(0, 'COUNT', count['CONVERT'])
count = count['COUNT']
count = pd.DataFrame(count)
```

```python
count = count[count.COUNT > 25]
#print count
#rcParams['figure.figsize'] = 40, 24
#count.plot(kind = 'bar')

#count['CLASS'] = np.nan
count.loc[count['COUNT']<6000, 'CLASS'] = 10
count.loc[count['COUNT']<4028, 'CLASS'] = 9
count.loc[count['COUNT']<895, 'CLASS'] = 8
count.loc[count['COUNT']<545, 'CLASS'] = 6
count.loc[count['COUNT']<370, 'CLASS'] = 6
count.loc[count['COUNT']<249, 'CLASS'] = 5
count.loc[count['COUNT']<180, 'CLASS'] = 4
count.loc[count['COUNT']<102, 'CLASS'] = 3
count.loc[count['COUNT']<60, 'CLASS'] = 2
count.loc[count['COUNT']<40, 'CLASS'] = 1

count.reset_index(level=0, inplace=True)



def merge_df(df,map,on):
    df = pd.DataFrame.merge(df,map,how='left',on=on)
    return df
```

```python
merged_audit = merge_df(audit, count, 'AUDITCONTRACTOR')
del merged_audit['COUNT']
del merged_audit['AUDITCONTRACTOR']
merged_audit.dropna()
merged_audit.to_csv('C:\Users\William\Documents\GitHub\urban_project_Haozhe\sample0722_HW_contractor_count.csv', index = False)

print merged_audit.head()
cols_to_keep = ['CONVERT', 'SQFT_LN', 'ANNUALPRIMARYUSAGE_LN', 'MEDIANHOUSEVALUE_LN']
#cols_to_keep = ['CONVERT', 'SQFT_LN', 'ANNUALPRIMARYUSAGE_LN']


#get dummies for categorical attributes
dummy_homeType = pd.get_dummies(audit['HOMETYPE'], prefix='HomeType')
dummy_income = pd.get_dummies(audit['INCOME'], prefix='Income')
dummy_primaryFuel = pd.get_dummies(audit['PRIMARYFUELTYPE'], prefix='Fuel')
dummy_region = pd.get_dummies(audit['REGION'], prefix='Region')
dummy_central = pd.get_dummies(audit['CENTRALAIR'], prefix='CentralAir')
dummy_owner = pd.get_dummies(audit['OWNER'], prefix='Owner')
dummy_contractor = pd.get_dummies(audit['AUDITCONTRACTOR'], prefix='Contractor')#this one need to be taken care of
# manually add the intercept
audit['intercept'] = 1.0
```

```python
pieces = [dummy_homeType[0:-1], dummy_income[0:-1], dummy_region[1:], dummy_primaryFuel[0:], dummy_contractor[0:], dummy_central[0:], dummy_owner[0:]]#one column was not inclded in region
data = merged_audit[cols_to_keep].join(pieces)
data = data.dropna()
#print data.head()
```

```python
train_cols = data.columns[1:]
data = data.dropna()
#print data.head()
data = data.dropna(subset=['CONVERT'], how='all')
```
